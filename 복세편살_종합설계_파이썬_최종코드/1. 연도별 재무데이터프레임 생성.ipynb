```python
import pandas as pd
import numpy as np

# 1. use fs_df.csv
fs_df = pd.read_csv('fs_df.csv', encoding='EUC-KR')
fs_df = fs_df.rename(columns={'Unnamed: 0':'ticker'})
fs_df['ticker'] = fs_df['ticker'].map(lambda x:'%06d'%x)

fs_df = fs_df.set_index('ticker')

#2. 연도별로 나누기
fs_2017_df = fs_df.iloc[:, np.arange(0, 948, 4)]
fs_2018_df = fs_df.iloc[:, np.arange(1, 948, 4)]
fs_2019_df = fs_df.iloc[:, np.arange(2, 948, 4)]
fs_2020_df = fs_df.iloc[:, np.arange(3, 948, 4)]

fs_2017_df.columns = [i.split('.')[0] for i in fs_2017_df.columns]
fs_2018_df.columns = [i.split('.')[0] for i in fs_2018_df.columns]
fs_2019_df.columns = [i.split('.')[0] for i in fs_2019_df.columns]
fs_2020_df.columns = [i.split('.')[0] for i in fs_2020_df.columns]

#3. SAVE 연도별 CSV
fs_2017_df.to_csv('data/fs_2017.csv')
fs_2018_df.to_csv('data/fs_2018.csv')
fs_2019_df.to_csv('data/fs_2019.csv')
fs_2020_df.to_csv('data/fs_2020.csv')
```
